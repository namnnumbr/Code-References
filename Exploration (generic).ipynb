## Blind Drops
Drop when all values missing or all values constant

```python
def blindDrop(df):
    """
    Returns updated df less columns where all values are missing or constant
    """
    missing = df.columns[df.isna().all()].tolist()
    if (len(missing) > 0):
        print('Cols all missing: ', missing)
        df = df.drop(columns=missing)
    else:
        print('No cols all missing')
        
    constant = df.columns[df.nunique()==1].tolist()        
    if (len(constant) > 0):
        print('Cols all constant: ', constant)
        df = df.drop(columns=constant)
    else:
        print('No cols all constant')
        
    return df
```

## Review Missingness

```python
def rmvMissing(df, thresh):
    """
    Identify columns where missingness exceeds threshold.
    Drop columns with specified degree of missingness.
    Return cleaned df.
    """
    
    if not isinstance(df, pd.DataFrame):
        raise Exception('df is not DataFrame.')

    if (thresh < 0 or thresh > 1):
        raise Exception('threshold represents % missingness and should be between 0 and 1 inclusive')
    
    nrow = df.shape[0]
    ncol = df.shape[1]
    
    # keep columns with missingness < threshold
    missing_cols = df.loc[:, ( (df.isna().sum(axis=0)/nrow) > thresh )].columns.tolist()
    df = df.drop(columns=missing_cols)

    print('Dropped columns: ', missing_cols)
    
    return df
```

```python
# categorical - fill with 'missing' as new category:
def addMissingCat(df, varnames):
    """
    Add category value 'missing' for missing data in specified columns
    """
    for col in df[varnames].select_dtypes(include=['category']).columns:
        df[col] = df[col].cat.add_categories('missing')
        df[col] = df[col].fillna('missing')

    return df
```

```python
# categorical - replace category name with numeric code
def catToCode(df):
    """
    Converts categorical variables to numeric identity
    """
    for col in df.select_dtypes(include=['category']).columns:
        df[col] = df[col].cat.codes
    
    # or
    # df = df.apply(lambda x : x.cat.codes if x.dtype == 'category' else x)
    
    return df
```
